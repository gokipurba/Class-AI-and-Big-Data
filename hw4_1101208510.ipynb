```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Downloads/Dengue_Daily_EN.csv")
df
```

```python
new_df = df[["Date_Confirmation","Sex", "Age_Group", "County_living"]]
new_df
```

```python
new_df = new_df.loc[new_df["Date_Confirmation"]!= "None"]
new_df.reset_index(drop = True, inplace = True)
new_df
```

```python
county = new_df.groupby(new_df["County_living"], sort = False)
county = dict(county["County_living"].count())
sortedByValue = {k:v for k, v in sorted(county.items(), key = lambda v: v[1], reverse = True)[:10]}
group_data = list(sortedByValue.values())
group_names = list(sortedByValue.keys())
```

```python
age = new_df.groupby(new_df["Age_Group"], sort = False)
age = dict(age["Age_Group"].count())
sortedByValueage = {k:v for k, v in sorted(age.items(), key = lambda v: v[1], reverse = True)[:10]}
group_adata = list(sortedByValueage.values())
group_anames = list(sortedByValueage.keys())
```

```python
new_df.dtypes
new_df['Date_Confirmation'] = pd.to_datetime(new_df.Date_Confirmation)
```

```python
new_df["month"]= new_df.Date_Confirmation.dt.month
new_df.month.value_counts().sort_index().plot()
fig, ax = plt.subplots(figsize = (8,5))
plt.title ("Age")
ax.barh(group_anames, group_adata)
fig, ax = plt.subplots(figsize = (8,5))
plt.title ("DBD")
ax.barh(group_names, group_data)
plt.show()
#sortedByValuedate = {k:v for k, v in sorted(month.items(), key = lambda v: v[1], reverse = True)[:10]}
#group_datae = list(sortedByValuedate.values())
#group_namese = list(sortedByValuedate.keys())
```
